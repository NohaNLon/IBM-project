# Decision Tree


The following lines of code fit a <code>DecisionTreeClassifier</code>:


```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
```

<b>Question  3</b> Determine the minumum   value for the parameter <code>max_depth</code> that improves results 


```python
# Define the range of max_depth values to test
max_depth_values = range(1, 21)  # for example, test max_depth from 1 to 20

# Initialize lists to store mean accuracy scores for each max_depth
mean_scores = []

# Iterate over max_depth values and evaluate the model using cross-validation
for max_depth in max_depth_values:

    decision_tree = DecisionTreeClassifier(criterion="entropy", max_depth = max_depth)
    # Perform cross-validation with 5 folds
    scores = cross_val_score(decision_tree, X_train, y_train, cv=5)
    
    # Calculate mean accuracy score and store it
    mean_score = scores.mean()
    mean_scores.append(mean_score)

# Find the index of the max mean accuracy score
best_max_depth_index = mean_scores.index(max(mean_scores))

# Get the corresponding best max_depth value
best_max_depth = max_depth_values[best_max_depth_index]

# Create the Decision Tree Classifier with the best max_depth value
best_decision_tree = DecisionTreeClassifier(criterion="entropy", max_depth=best_max_depth)

# Fit the classifier to the training data
best_decision_tree.fit(X_train, y_train)

# Evaluate the model on the test set
accuracy = best_decision_tree.score(X_val, y_val)
print(f"The minimum max_depth that improves results is: {best_max_depth} with an accuracy of {accuracy}")
```

# Support Vector Machine


<b>Question  4</b> Train the support  vector machine model and determine the accuracy on the validation data for each kernel. Find the kernel (linear, poly, rbf, sigmoid) that provides the best score on the validation data and train a SVM using it.


```python
from sklearn import svm
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score

# Define the range of C values to test
C_values = [0.1, 1, 10, 100]  # Example values, you can adjust as needed

# Initialize lists to store mean accuracy scores for each kernel
mean_scores_linear = []
mean_scores_poly = []
mean_scores_rbf = []
mean_scores_sigmoid = []

# Iterate over C values and evaluate the models using cross-validation
for C in C_values:
    # Linear kernel
    svm_linear = SVC(kernel='linear', C=C)
    scores_linear = cross_val_score(svm_linear, X_train, y_train, cv=5)
    mean_scores_linear.append(scores_linear.mean())
    
    # Polynomial kernel
    svm_poly = SVC(kernel='poly', C=C)
    scores_poly = cross_val_score(svm_poly, X_train, y_train, cv=5)
    mean_scores_poly.append(scores_poly.mean())
    
    # RBF kernel
    svm_rbf = SVC(kernel='rbf', C=C)
    scores_rbf = cross_val_score(svm_rbf, X_train, y_train, cv=5)
    mean_scores_rbf.append(scores_rbf.mean())
    
    # Sigmoid kernel
    svm_sigmoid = SVC(kernel='sigmoid', C=C)
    scores_sigmoid = cross_val_score(svm_sigmoid, X_train, y_train, cv=5)
    mean_scores_sigmoid.append(scores_sigmoid.mean())

# Find the index of the max mean accuracy score for each kernel
best_linear_index = mean_scores_linear.index(max(mean_scores_linear))
best_poly_index = mean_scores_poly.index(max(mean_scores_poly))
best_rbf_index = mean_scores_rbf.index(max(mean_scores_rbf))
best_sigmoid_index = mean_scores_sigmoid.index(max(mean_scores_sigmoid))

# Get the corresponding best C value for each kernel
best_C_linear = C_values[best_linear_index]
best_C_poly = C_values[best_poly_index]
best_C_rbf = C_values[best_rbf_index]
best_C_sigmoid = C_values[best_sigmoid_index]

# Choose the kernel with the highest mean accuracy score on the validation data
best_kernel = max([(mean_scores_linear[best_linear_index], 'linear'),
                   (mean_scores_poly[best_poly_index], 'poly'),
                   (mean_scores_rbf[best_rbf_index], 'rbf'),
                   (mean_scores_sigmoid[best_sigmoid_index], 'sigmoid')])[1]

# Train a new SVM model using the best kernel and C value
if best_kernel == 'linear':
    best_svm = SVC(kernel='linear', C=best_C_linear)
elif best_kernel == 'poly':
    best_svm = SVC(kernel='poly', C=best_C_poly)
elif best_kernel == 'rbf':
    best_svm = SVC(kernel='rbf', C=best_C_rbf)
else:
    best_svm = SVC(kernel='sigmoid', C=best_C_sigmoid)

# Fit the classifier to the training data
best_svm.fit(X_train, y_train)

# Evaluate the model on the validation set
accuracy = best_svm.score(X_val, y_val)
print(f"The best kernel is {best_kernel} with C={getattr(best_svm, 'C', None)} and an accuracy of {accuracy}")
```

```python
import numpy as np

def jaccard_score_custom(y_true, y_pred):
    try:
        # Ensure y_true and y_pred are numpy arrays
        y_true = np.array(y_true)
        y_pred = np.array(y_pred)

        # Check if y_true and y_pred have the same shape
        if y_true.shape != y_pred.shape:
            raise ValueError("y_true and y_pred must have the same shape")

        # Check if y_true and y_pred contain valid labels
        if not np.all(np.isin(y_true, [0, 1])) or not np.all(np.isin(y_pred, [0, 1])):
            raise ValueError("y_true and y_pred must contain binary labels (0 or 1)")

        # Compute Jaccard score for binary classification
        intersection = np.logical_and(y_true, y_pred)
        union = np.logical_or(y_true, y_pred)
        jaccard = np.sum(intersection) / np.sum(union)

        return jaccard

    except Exception as e:
        print("Error:", e)
        return None

# Example usage:
y_true = [0, 1, 1, 0, 1]
y_pred = [1, 1, 0, 0, 1]
print("Jaccard Score:", jaccard_score_custom(y_true, y_pred))

from sklearn.metrics import f1_score

# Assuming y_true and y_pred are your true and predicted labels, respectively
y_true = [0, 1, 1, 0, 1]
y_pred = [1, 1, 0, 0, 1]

# Calculate F1 score
f1 = f1_score(y_true, y_pred, average='micro')

print("F1 Score:", f1)




```
